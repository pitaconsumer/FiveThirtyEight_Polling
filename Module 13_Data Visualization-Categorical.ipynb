##Part 1, Q2: Choose two continuous variables, and plot them three different ways.

```python
#Selecting "adjusted_approve" and "adjusted_disaprove" || Note categorical is "grade"
g = sns.lmplot(y='adjusted_approve', # Variable 1.
               x='adjusted_disapprove', # Variable 2.
               data=df, # Data
               fit_reg=False, # If set to true, plots a regression line.
               scatter_kws={'alpha':0.4}) # Set points to semi-transparent to see overlapping points.
g.set_ylabels("Approve")
g.set_xlabels("Disapprove")
plt.title('Plot1: Scatterplot of Drumpf Disapproval')
plt.show()

g = sns.lmplot(y='adjusted_disapprove', 
               x='adjusted_approve',
               data=df, # Data.
               fit_reg=True, # The regression line also includes a 95% confidence envelope.
               scatter_kws={'alpha':0.4})
g.set_ylabels("adjusted_disapprove")
g.set_xlabels("adjusted_approve")
plt.title('Plot 2: Scatterplot with regression line: Drumpf Disapproval')
plt.show()
```

##Jointplot: Second Way from Seaboorns

```python
mean, cov = [0, 1], [(1, .5), (.5, 1)]
data = np.random.multivariate_normal(mean, cov, 200)
df = pd.DataFrame(data, columns=["pollster", "approve"])
sns.jointplot(x="pollster", y="approve", data=df)
```
